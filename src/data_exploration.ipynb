```python
import pandas as pd

df_2008 = pd.read_csv("../data/atp_matches_2008.csv")
df_2009 = pd.read_csv("../data/atp_matches_2009.csv")
df_2010 = pd.read_csv("../data/atp_matches_2010.csv")
df_2011 = pd.read_csv("../data/atp_matches_2011.csv")
df_2012 = pd.read_csv("../data/atp_matches_2012.csv")
df_2013 = pd.read_csv("../data/atp_matches_2013.csv")
df_2014 = pd.read_csv("../data/atp_matches_2014.csv")
df_2015 = pd.read_csv("../data/atp_matches_2015.csv")
df_2016 = pd.read_csv("../data/atp_matches_2016.csv")
df_2017 = pd.read_csv("../data/atp_matches_2017.csv")
df_2018 = pd.read_csv("../data/atp_matches_2018.csv")
df_2019 = pd.read_csv("../data/atp_matches_2019.csv")
df_2020 = pd.read_csv("../data/atp_matches_2020.csv")
df_2021 = pd.read_csv("../data/atp_matches_2021.csv")
df_2022 = pd.read_csv("../data/atp_matches_2022.csv")
df_2023 = pd.read_csv("../data/atp_matches_2023.csv")
df_2024 = pd.read_csv("../data/atp_matches_2024.csv")
df_2025 = pd.read_csv("../data/atp_matches_2025.csv")
df_2026 = pd.read_csv("../data/atp_matches_2026.csv")

df_pre_model = pd.concat([df_2008,df_2009,df_2010,df_2011,df_2012,df_2013,df_2014,df_2015,df_2016,df_2017,df_2018,df_2019,df_2020,df_2021,df_2022,df_2023,df_2024,df_2025,df_2026],ignore_index=True)

df_pre_model
```

```python


df_pre_model["matchup_id"] = df_pre_model.apply(lambda r: f"{min(r['winner_id'], r['loser_id'])}_{max(r['winner_id'], r['loser_id'])}", axis=1)
df_pre_model
```

```python
# 1. Extrai o ID do p1 (tudo antes do '_') diretamente da coluna de forma vetorizada
# Convertemos para string para garantir que a comparação funcione
p1_ids = df_pre_model['matchup_id'].str.split('_').str[0]

df_pre_model["p1_won"] = (df_pre_model['winner_id'].astype(str) == p1_ids).astype(int)

df_pre_model['p2_won'] = 1 - df_pre_model['p1_won']
```

```python
# 1. Quantidade de partidas que eles já jogaram ANTES da partida atual
# O .cumcount() conta as linhas do grupo (0, 1, 2...). 
# Como a primeira linha vira 0, ele já funciona nativamente como o "passado", sem precisar de shift!
df_pre_model['total_h2h_matches'] = df_pre_model.groupby('matchup_id').cumcount()

# 2. Vitórias do Player 1 ANTES da partida atual (usando transform para não vazar o grupo)
df_pre_model['p1_h2h_wins'] = df_pre_model.groupby('matchup_id')['p1_won'].transform(lambda x: x.shift(1).cumsum()).fillna(0)

# 3. Vitórias do Player 2 ANTES da partida atual
df_pre_model['p2_h2h_wins'] = df_pre_model['total_h2h_matches'] - df_pre_model['p1_h2h_wins']
```

```python
# 1. Total de partidas que eles jogaram antes NO PISO ATUAL da linha
df_pre_model['total_h2h_matches_surface'] = df_pre_model.groupby(['matchup_id', 'surface']).cumcount()

# 2. Vitórias do Player 1 antes NO PISO ATUAL da linha (usando transform para isolar o grupo composto)
df_pre_model['p1_h2h_wins_surface'] = df_pre_model.groupby(['matchup_id', 'surface'])['p1_won'].transform(lambda x: x.shift(1).cumsum()).fillna(0)

# 3. Vitórias do Player 2 antes NO PISO ATUAL
df_pre_model['p2_h2h_wins_surface'] = df_pre_model['total_h2h_matches_surface'] - df_pre_model['p1_h2h_wins_surface']
```

```python
# Certifique-se de que a base está ordenada por data antes de rodar
df_pre_model = df_pre_model.sort_values('tourney_date').reset_index(drop=True)

# 1. Criamos os dados deslocados (shift) para garantir que o jogo de hoje não entre no cálculo
df_pre_model['p1_won_shifted'] = df_pre_model.groupby('matchup_id')['p1_won'].shift(1)

# Criamos um marcador de "1" para cada partida ocorrida, para contar quantas partidas aconteceram na janela
df_pre_model['match_marker_shifted'] = df_pre_model.groupby('matchup_id')['p1_won'].transform(lambda x: x.index * 0 + 1).shift(1)

# 2. Calculamos o total de partidas jogadas na janela das últimas 10
df_pre_model['total_h2h_matches_last_10'] = (
    df_pre_model.groupby('matchup_id')['match_marker_shifted']
    .rolling(window=10, min_periods=1)
    .sum()
    .reset_index(level=0, drop=True)
    .fillna(0)
)

# 3. Calculamos quantas dessas últimas 10 o Player 1 venceu
df_pre_model['p1_h2h_wins_last_10'] = (
    df_pre_model.groupby('matchup_id')['p1_won_shifted']
    .rolling(window=10, min_periods=1)
    .sum()
    .reset_index(level=0, drop=True)
    .fillna(0)
)

# 4. Por consequência, as vitórias do Player 2 são a diferença
df_pre_model['p2_h2h_wins_last_10'] = df_pre_model['total_h2h_matches_last_10'] - df_pre_model['p1_h2h_wins_last_10']

# Opcional: Limpar as colunas auxiliares de shift para não poluir o dataset
df_pre_model = df_pre_model.drop(columns=['p1_won_shifted', 'match_marker_shifted'])

df_pre_model
```

```python
# Certifique-se de que a base está ordenada por data
df_pre_model = df_pre_model.sort_values('tourney_date').reset_index(drop=True)

# 1. Criamos os dados deslocados (shift) por matchup e superfície
df_pre_model['p1_won_shifted_surf'] = df_pre_model.groupby(['matchup_id', 'surface'])['p1_won'].shift(1)
df_pre_model['match_marker_shifted_surf'] = df_pre_model.groupby(['matchup_id', 'surface'])['p1_won'].transform(lambda x: x.index * 0 + 1).shift(1)

# 2. Total de partidas entre eles nas últimas 5 QUE FORAM NESSE PISO
df_pre_model['total_h2h_matches_last_10_surf'] = (
    df_pre_model.groupby(['matchup_id', 'surface'])['match_marker_shifted_surf']
    .rolling(window=10, min_periods=1)
    .sum()
    .reset_index(level=[0, 1], drop=True) # Remove os dois níveis do MultiIndex (matchup e surface)
    .fillna(0)
)

# 3. Vitórias do Player 1 nas últimas 10 QUE FORAM NESSE PISO
df_pre_model['p1_h2h_wins_last_10_surf'] = (
    df_pre_model.groupby(['matchup_id', 'surface'])['p1_won_shifted_surf']
    .rolling(window=10, min_periods=1)
    .sum()
    .reset_index(level=[0, 1], drop=True)
    .fillna(0)
)

# 4. Vitórias do Player 2
df_pre_model['p2_h2h_wins_last_10_surf'] = df_pre_model['total_h2h_matches_last_10_surf'] - df_pre_model['p1_h2h_wins_last_10_surf']

# Limpa colunas auxiliares
df_pre_model = df_pre_model.drop(columns=['p1_won_shifted_surf', 'match_marker_shifted_surf'])
```

```python
import pandas as pd
import numpy as np

# Garanta que a base original está ordenada por data
df_pre_model = df_pre_model.sort_values('tourney_date').reset_index(drop=True)

# 1. Isolamos as estatísticas de quem VENCEU
vencedores = df_pre_model[[
    'tourney_date', 'winner_id', 'w_ace', 'w_df', 'w_svpt', 
    'w_1stIn', 'w_1stWon', 'w_2ndWon', 'w_SvGms', 'w_bpSaved', 'w_bpFaced'
]].rename(columns={
    'winner_id': 'player_id', 'w_ace': 'ace', 'w_df': 'df', 'w_svpt': 'svpt',
    'w_1stIn': 'first_in', 'w_1stWon': 'first_won', 'w_2ndWon': 'second_won',
    'w_SvGms': 'sv_gms', 'w_bpSaved': 'bp_saved', 'w_bpFaced': 'bp_faced'
})

# 2. Isolamos as estatísticas de quem PERDEU
perdedores = df_pre_model[[
    'tourney_date', 'loser_id', 'l_ace', 'l_df', 'l_svpt', 
    'l_1stIn', 'l_1stWon', 'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced'
]].rename(columns={
    'loser_id': 'player_id', 'l_ace': 'ace', 'l_df': 'df', 'l_svpt': 'svpt',
    'l_1stIn': 'first_in', 'l_1stWon': 'first_won', 'l_2ndWon': 'second_won',
    'l_SvGms': 'sv_gms', 'l_bpSaved': 'bp_saved', 'l_bpFaced': 'bp_faced'
})

# 3. Juntamos as duas tabelas e ordenamos por data
player_perf = pd.concat([vencedores, perdedores]).sort_values('tourney_date').reset_index(drop=True)
```

```python
# Lista de colunas que vamos calcular a média
colunas_stats = ['ace', 'df', 'svpt', 'first_in', 'first_won', 'second_won', 'sv_gms', 'bp_saved', 'bp_faced']

# Criamos um DataFrame temporário para guardar as médias
medias_temporarias = pd.DataFrame(index=player_perf.index)

# Loop rápido apenas para aplicar o cálculo em cada coluna estatística
for col in colunas_stats:
    medias_temporarias[f'avg_{col}_last_15'] = (
        player_perf.groupby('player_id')[col]
        .shift(1) # Empurra para baixo (anti-vazamento)
        .rolling(window=15, min_periods=1) # Janela das últimas 15 partidas
        .mean()
        .reset_index(level=0, drop=True)
    )

# Adicionamos as médias de volta na tabela de performance dos jogadores
player_perf_final = pd.concat([player_perf[['tourney_date', 'player_id']], medias_temporarias], axis=1).fillna(0)
```

```python
# Descobrimos quem é o p1_id e p2_id em cada linha da base original
df_pre_model['p1_id'] = df_pre_model[['winner_id', 'loser_id']].min(axis=1)
df_pre_model['p2_id'] = df_pre_model[['winner_id', 'loser_id']].max(axis=1)

# 1. Merge para o Player 1
# O drop_duplicates garante que se ele jogou duas vezes no mesmo dia (raro, mas acontece), não duplique linhas
player_perf_final_p1 = player_perf_final.drop_duplicates(subset=['tourney_date', 'player_id'])
df_pre_model = pd.merge(
    df_pre_model, 
    player_perf_final_p1.add_prefix('p1_'), 
    left_on=['tourney_date', 'p1_id'], 
    right_on=['p1_tourney_date', 'p1_player_id'], 
    how='left'
).drop(columns=['p1_tourney_date', 'p1_player_id'])

# 2. Merge para o Player 2
player_perf_final_p2 = player_perf_final.drop_duplicates(subset=['tourney_date', 'player_id'])
df_pre_model = pd.merge(
    df_pre_model, 
    player_perf_final_p2.add_prefix('p2_'), 
    left_on=['tourney_date', 'p2_id'], 
    right_on=['p2_tourney_date', 'p2_player_id'], 
    how='left'
).drop(columns=['p2_tourney_date', 'p2_player_id'])

# Preenche eventuais vazios (primeiros jogos da história do jogador) com 0
colunas_novas = [f'p1_avg_{c}_last_15' for c in colunas_stats] + [f'p2_avg_{c}_last_15' for c in colunas_stats]
df_pre_model[colunas_novas] = df_pre_model[colunas_novas].fillna(0)
```

```python
# Convertemos a coluna para o formato de data do Pandas (caso já não esteja)
df_pre_model['tourney_date'] = pd.to_datetime(df_pre_model['tourney_date'], format='%Y%m%d')

# Filtramos a base: guardamos apenas as partidas de 2011 em diante para treinar o modelo
df_final_para_modelo = df_pre_model[df_pre_model['tourney_date'].dt.year >= 2011].reset_index(drop=True)
```

```python
df_final_para_modelo['tourney_date'] = pd.to_datetime(df_final_para_modelo['tourney_date'])
df_treino = df_final_para_modelo[df_final_para_modelo['tourney_date'].dt.year >= 2011].copy()
# --- 1. SEPARAÇÃO DAS DUAS METADES (ESPELHAMENTO COMPLETO) ---

# Metade 1: P1 é o Vencedor (Target = 1)
metade_vitoria = pd.DataFrame()
metade_vitoria['target'] = np.ones(len(df_treino))

# Características de Contexto/Perfil (P1 = Vencedor, P2 = Perdedor)
metade_vitoria['p1_age'] = df_treino['winner_age']
metade_vitoria['p2_age'] = df_treino['loser_age']
metade_vitoria['p1_rank'] = df_treino['winner_rank']
metade_vitoria['p2_rank'] = df_treino['loser_rank']
metade_vitoria['p1_rank_pts'] = df_treino['winner_rank_points']
metade_vitoria['p2_rank_pts'] = df_treino['loser_rank_points']

# Histórico Total de Confrontos (H2H)
metade_vitoria['p1_h2h_wins_all'] = df_treino['p1_h2h_wins']
metade_vitoria['p2_h2h_wins_all'] = df_treino['p2_h2h_wins']
metade_vitoria['p1_h2h_wins_surf'] = df_treino['p1_h2h_wins_surface']
metade_vitoria['p2_h2h_wins_surf'] = df_treino['p2_h2h_wins_surface']

# Histórico Recente de Confrontos (Últimas 10 no piso - que calculamos depois)
metade_vitoria['p1_h2h_wins_recent_surf'] = df_treino['p1_h2h_wins_last_10_surf']
metade_vitoria['p2_h2h_wins_recent_surf'] = df_treino['p2_h2h_wins_last_10_surf']

# Médias Móveis de Performance (Últimas 15 partidas)
for col in colunas_stats:
    metade_vitoria[f'p1_avg_{col}'] = df_treino[f'p1_avg_{col}_last_15']
    metade_vitoria[f'p2_avg_{col}'] = df_treino[f'p2_avg_{col}_last_15']


# Metade 2: P1 é o Perdedor (Target = 0) - INVERTENDO TUDO
metade_derrota = pd.DataFrame()
metade_derrota['target'] = np.zeros(len(df_treino))

# Características de Contexto/Perfil (Invertido)
metade_derrota['p1_age'] = df_treino['loser_age']
metade_derrota['p2_age'] = df_treino['winner_age']
metade_derrota['p1_rank'] = df_treino['loser_rank']
metade_derrota['p2_rank'] = df_treino['winner_rank']
metade_derrota['p1_rank_pts'] = df_treino['loser_rank_points']
metade_derrota['p2_rank_pts'] = df_treino['winner_rank_points']

# Histórico Total de Confrontos (Invertido)
metade_derrota['p1_h2h_wins_all'] = df_treino['p2_h2h_wins']
metade_derrota['p2_h2h_wins_all'] = df_treino['p1_h2h_wins']
metade_derrota['p1_h2h_wins_surf'] = df_treino['p2_h2h_wins_surface']
metade_derrota['p2_h2h_wins_surf'] = df_treino['p1_h2h_wins_surface']

# Histórico Recente de Confrontos (Invertido)
metade_derrota['p1_h2h_wins_recent_surf'] = df_treino['p2_h2h_wins_last_10_surf']
metade_derrota['p2_h2h_wins_recent_surf'] = df_treino['p1_h2h_wins_last_10_surf']

# Médias Móveis de Performance (Invertido)
for col in colunas_stats:
    metade_derrota[f'p1_avg_{col}'] = df_treino[f'p2_avg_{col}_last_15']
    metade_derrota[f'p2_avg_{col}'] = df_treino[f'p1_avg_{col}_last_15']


# --- 2. JUNTAR AS DUAS METADES ---
base_modelagem = pd.concat([metade_vitoria, metade_derrota]).reset_index(drop=True)

# Trata possíveis rankings vazios (como jogadores novos que entram sem rank, coloca um valor alto tipo 999)
base_modelagem[['p1_rank', 'p2_rank']] = base_modelagem[['p1_rank', 'p2_rank']].fillna(999)
base_modelagem = base_modelagem.fillna(0)


# --- 3. AUTOMAÇÃO DA CRIAÇÃO DE TODAS AS DIFERENÇAS (O "PULO DO GATO") ---
# Em vez de fazer uma por uma, vamos pegar todos os pares de colunas p1_ e p2_ e calcular a diferença

# Lista com os nomes base de todas as características
recursos_base = ['age', 'rank', 'rank_pts', 'h2h_wins_all', 'h2h_wins_surf', 'h2h_wins_recent_surf'] + [f'avg_{col}' for col in colunas_stats]

for recurso in recursos_base:
    base_modelagem[f'diff_{recurso}'] = base_modelagem[f'p1_{recurso}'] - base_modelagem[f'p2_{recurso}']


# --- 4. SEPARAÇÃO FINAL (X e y) ---
y = base_modelagem['target']
X = base_modelagem.drop(columns=['target'])

print(f"Base de dados finalizada!")
print(f"Quantidade total de colunas de recursos (features) para o modelo: {X.shape[1]}")
```
